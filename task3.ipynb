```python
from transformers import MBartForConditionalGeneration,MBart50TokenizerFast
```

```python
model= MBartForConditionalGeneration.from_pretrained('facebook/mbart-large-50-one-to-many-mmt')
```

```python
tokenizer = MBart50TokenizerFast.from_pretrained('facebook/mbart-large-50-one-to-many-mmt',src_lang='en_XX')
```

```python
input_text=['I had about a 30 minute demo just using this new headset']
```

```python
model_input=tokenizer(input_text,return_tensors='pt',padding=True,truncation=True)
```

```python

# translate english to hindi

generated_tokens = model.generate(**model_input,
    forced_bos_token_id=tokenizer.lang_code_to_id['hi_IN']
)
```

```python
traslate = tokenizer.batch_decode(generated_tokens,skip_special_tokens=True)
```
